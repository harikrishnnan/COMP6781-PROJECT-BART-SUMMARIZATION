```python
def calc_rouge_scores(score_list,score_list_bert,score_list_bleu,string,dataset,addIndex):
    rouge1=0
    rouge1min=99
    rouge1max=0
    rouge2=0
    rouge2min=99
    rouge2max=0
    rougeLsum=0
    rougeLsummin=99
    rougeLsummax=0
    bert_f1_sum=0
    bert_f1_min=99
    bert_f1_max=0
    bleu=0
    bleu_min=99
    bleu_max=0
    for index,i in enumerate(score_list):
        
        rougeLsum+=i["rougeLsum"]
        if i["rougeLsum"]>rougeLsummax:
            rougeLsummax=i["rougeLsum"]
        elif i["rougeLsum"]<rougeLsummin:
            rougeLsummin=i["rougeLsum"]
        
        rouge2+=i["rouge2"]
        if i["rouge2"]>rouge2max:
            rouge2max=i["rouge2"]
        elif i["rouge2"]<rouge2min:
            rouge2min=i["rouge2"]
            
        rouge1+=i["rouge1"]
        if i["rouge1"]>rouge1max:
            rouge1max=i["rouge1"]
        elif i["rouge1"]<rouge1min:
            rouge1min=i["rouge1"]
            
        bert_f1_sum+=score_list_bert[index+addIndex]["f1"][0]
        if score_list_bert[index+addIndex]["f1"][0]>bert_f1_max:
            bert_f1_max=score_list_bert[index+addIndex]["f1"][0]
        elif score_list_bert[index+addIndex]["f1"][0]<bert_f1_min:
            bert_f1_min=score_list_bert[index+addIndex]["f1"][0]
            
        bleu+=score_list_bleu[index+addIndex]["google_bleu"]
        if score_list_bleu[index+addIndex]["google_bleu"]>bleu_max:
            bleu_max=score_list_bleu[index+addIndex]["google_bleu"]
        elif score_list_bleu[index+addIndex]["google_bleu"]<bleu_min:
            bleu_min=score_list_bleu[index+addIndex]["google_bleu"]
            
    rouge1=rouge1/1000
    rouge2=rouge2/1000
    rougeLsum=rougeLsum/1000
    bert_f1_sum=bert_f1_sum/1000
    bleu=bleu/1000
    print(f"Rouge-1 score {string} - {dataset}",rouge1*100)
    print(f"Rouge-1 score min {string}- {dataset}",rouge1min*100)
    print(f"Rouge-1 score max {string}- {dataset}",rouge1max*100)
    print(f"Rouge-2 score  {string}- {dataset}",rouge2*100)
    print(f"Rouge-2 score min {string}- {dataset}",rouge2min*100)
    print(f"Rouge-2 score max {string}- {dataset}",rouge2max*100)
    print(f"Rouge-Lsum score  {string}- {dataset}",rougeLsum*100)
    print(f"Rouge-Lsum score min {string}- {dataset}",rougeLsummin*100)
    print(f"Rouge-Lsum score max {string}- {dataset}",rougeLsummax*100)
    
    print(f"BERT F1 score {string}- {dataset}",bert_f1_sum*100)
    print(f"BERT F1 score min {string}- {dataset}",bert_f1_min*100)
    print(f"BERT F1 score max {string}- {dataset}",bert_f1_max*100)
    print(f"BLEU score {string}- {dataset}",bleu*100)
    print(f"BLEU score min{string}- {dataset}",bleu_min*100)
    print(f"BLEU score max{string}- {dataset}",bleu_max*100)
```
